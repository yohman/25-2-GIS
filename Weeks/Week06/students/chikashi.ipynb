```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
reitaku = [35.8330, 139.9550]

kaede = [35.83304899572845, 139.9550263688704]
asunaro = [35.83218389343381, 139.9559046419524]
hiiragi = [35.83432746243043, 139.95492044150396]
library = [35.83288620720931, 139.9560383893833]
parking = [35.83477548217266, 139.9561361979999]
bukastu = [35.83480867733186, 139.95489801198133]


seven = [35.83153177619921, 139.95615584329036]
moscafe = [35.831142717397036, 139.95596014665873]
primary = [35.83637864291207, 139.95583678516158]
police = [35.83038280781122, 139.95595654760183]



# make the map
m = folium.Map(location=reitaku, zoom_start=16)

# add a marker
folium.Marker(kaede, popup="麗澤大学のキャンパス", tooltip="かえで").add_to(m)
folium.Marker(asunaro, popup="麗澤大学のキャンパス", tooltip="あすなろ").add_to(m)
folium.Marker(hiiragi, popup="麗澤大学の学食", tooltip="ひいらぎ").add_to(m)
folium.Marker(library, popup="麗澤大学の図書館", tooltip="図書館").add_to(m)
folium.Marker(parking, popup="麗澤大学の駐車場", tooltip="駐車場").add_to(m)
folium.Marker(bukastu, popup="麗澤大学の部室", tooltip="部室棟").add_to(m)
folium.Marker(seven, popup="麗澤大学の近くのコンビニ", tooltip="セブンイレブン").add_to(m)
folium.Marker(moscafe, popup="麗澤大学の近くのカフェ", tooltip="モスカフェ").add_to(m)
folium.Marker(primary, popup="麗澤大学の近くの小学校", tooltip="光が丘小学校").add_to(m)
folium.Marker(police, popup="麗澤大学は柏市の近くの交番", tooltip="交番").add_to(m)

# show the map

m
```
