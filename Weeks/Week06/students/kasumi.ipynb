### 楽しかったところ

```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
disneysea = [35.62667987286921, 139.88508579276834]
disneyland = [35.63287023988384, 139.8804040054496]
usj = [34.66540172347249, 135.43235821374634]
domecityattractions = [35.70438120236655, 139.75339096895556]

# make the map
m = folium.Map(location=usj, zoom_start=8)

# add a marker
folium.Marker(usj, popup="ユニバーサルスタジオジャパン", tooltip="usj").add_to(m)
folium.Marker(disneysea, popup="ディズニーシー", tooltip="disneysea").add_to(m)
folium.Marker(disneyland, popup="ディズニーランド", tooltip="disneyland").add_to(m)
folium.Marker(domecityattractions, popup="屋内アトラクションがあるところ", tooltip="domecityattractions").add_to(m)

# show the map
m
```
