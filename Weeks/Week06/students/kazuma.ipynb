麗澤大学近くにある飲食店




(昼飯)(距離順)






```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
food1=[35.8343055, 139.9549949]
food2=[35.83225354489094, 139.95683088844197]
food3=[35.82998796737419, 139.95587968818734]
food4=[35.828105543972605, 139.95726465236046]
food5=[35.83515765132268, 139.946063855811]
# make the map
m = folium.Map(location=food1, zoom_start=15)

# add a marker
folium.Marker(food1, popup="学生会館食堂 棟「ひいらぎ」", tooltip="学生会館食堂 棟「ひいらぎ」").add_to(m)
folium.Marker(food2, popup="モスバーガー麗澤大学前店", tooltip="モスバーガー麗澤大学前店").add_to(m)
folium.Marker(food3, popup="中華菜館", tooltip="中華菜館").add_to(m)
folium.Marker(food4, popup="アジアンレストラン アビラル", tooltip="アジアンレストラン アビラル").add_to(m)
folium.Marker(food5, popup="激旨麺厨房 劉", tooltip="激旨麺厨房 劉").add_to(m)
# show the map
m
```
