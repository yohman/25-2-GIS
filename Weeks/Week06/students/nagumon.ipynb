第１位：ディズニーシー
アトラクションがたくさんで楽しいです！ディズニーランドより人気がある！（私の界隈では・・・。）

```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
disney_sea = [35.62586244524468, 139.88478754553358]
pulo = [35.624608693553384, 139.4290828599274]
ski = [37.047563339390344, 138.738682495039]
spot = [36.99463657458056, 138.7357459888181]#清津峡
skytree = [35.71007802321593, 139.81059852964788]
japan = [36.26584330938964, 138.72729570431514]#地図の中心地

# make the map
m = folium.Map(location=japan, zoom_start=8)

# add a marker
folium.Marker(disney_sea, popup="ディズニーシーだよ！楽しいよ！", tooltip="第１位").add_to(m)
folium.Marker(pulo,popup="キティーは好き？？",tooltip="第５位").add_to(m)
folium.Marker(ski,popup="スノボ楽しいよね！",tooltip="第4位").add_to(m)
folium.Marker(spot,popup="行ったことある？",tooltip="第3位").add_to(m)
folium.Marker(skytree,popup="高所恐怖症です汗",tooltip="第2位").add_to(m)

# show the map
m
```
