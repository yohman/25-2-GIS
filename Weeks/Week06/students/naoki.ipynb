# My map

I'll post some gym recommendations near here!
The green marker is Reitaku.

The red marker is the gym I use most often.

The purple marker is the closest gym to the University.
This gym is called Anytime and is located all over the world.

The orange marker is the gym I recommend the most.
He owner of this gym is an active physique athlete and it is a cheap, high quality gym.

```python
import folium

reitaku=(35.8330, 139.9550)
gym=folium.Map(location=reitaku,zoom_start=12)

folium.Marker(location=reitaku,
              popup='My University',
              tooltip='Reitaku',
              icon=folium.Icon(color="green")).add_to(gym)

mygym=35.8793487859615, 139.96800873757294
folium.Marker(location=mygym,popup='I always use the gym.',tooltip='My gym',icon=folium.Icon(color="red")).add_to(gym)

closegym=35.848318520795374, 139.95377511849304
folium.Marker(location=closegym,popup='These gyms are located all over the world.',tooltip='Gym close to the university',icon=folium.Icon(color="purple")).add_to(gym)

bestgym=35.82528901933755, 139.91992856931068
folium.Marker(location=bestgym,popup='This is the best gym, cheap and good quality.',tooltip='My recommendation for gyms.',icon=folium.Icon(color="orange")).add_to(gym)

gym
```
