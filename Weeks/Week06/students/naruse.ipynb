The important places for hip hop

```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
origin = [40.847993277156185, -73.92496867233002]
ie = [36.073282620729465, 140.20016109893683]
compton = [33.896076324668414, -118.22196201920562]
atl = [33.75469926551453, -84.40367214983806]

# make the map
m = folium.Map(location=origin,  zoom_start=15)


# add a marker
folium.Marker(origin, popup="The place where hip hop was born", tooltip="1520 Sedgwick Avenue").add_to(m)
folium.Marker(ie, popup="The place where the gratest hip-hoper lives", tooltip="My House").add_to(m)
folium.Marker(compton, popup="The place where Kendrick Lamar was born", tooltip="Compton").add_to(m)
folium.Marker(atl, popup="The third biggest place of hiphop", tooltip="Atlanta").add_to(m)
# show the map
m
```
