```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
a = [35.8621498,139.9709172]
b = [35.8464847,139.9546245]
c = [35.8716816,139.9879859]
d = [35.8411074,139.9726855]

# make the map
m = folium.Map(location=a, zoom_start=13)

# add a marker
folium.Marker(a, pupup = "tokai",tooltip="kashiwaeki").add_to(m)
folium.Marker(b, popup = "tokaijanai" , tooltip="minamikashiwaeki").add_to(m)
folium.Marker(c, popup = "tokaijanai1", tooltip="kitakashiwaeki").add_to(m)
folium.Marker(d, popup = "tokaijanai2", tooltip="shinkashiwaeki").add_to(m)
# show the map
m
```
