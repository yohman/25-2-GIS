# Mapping world data

https://www.gapminder.org/data/

```python
import pandas as pd
```

```python
df = pd.read_csv('data/hapiscore_whr.csv')
```

```python
df.head()
```

```python
df = df.melt(id_vars=["country"], 
        var_name="year", 
        value_name="happiness")
```

```python
df[df['country']=='Japan']
```

```python
df.query("country=='Japan'")
```

```python
df.query("country=='Japan'").plot.bar(x='year',y='happiness',color='red', title='Japan')
```

```python
df[df['country']=='South Korea'].plot.bar(x='year',y='happiness',color='navy',title='South Korea')
```

```python
def happiness(country):
    df[df['country']==country].plot.bar(x='year',y='happiness',color='navy', title=country)
```

```python
happiness('Philippines')
```

```python
countries = ['Japan','South Korea','Vietnam','Philippines']
```

```python
for country in countries:
    happiness(country)
```

## Merge


でもjoinするfieldnameが違う😩

`chiba_ku`は`N03_007`で、`chiba_pop`は`KEY_CODE`

```python
# rename
chiba_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
chiba_ku
```

さらにdatatypeも違う😩

```python
chiba_ku['KEY_CODE'] = chiba_ku['KEY_CODE'].astype(int)
```

```python
chiba_ku.info
```

```python
chiba_ku_pop = chiba_ku.merge(chiba_pop,on='KEY_CODE')
```

```python
chiba_ku_pop
```

# Make a choropleth map

chiba_ku_pop.plot()

```python
chiba_ku_pop['人口総数'] = chiba_ku_pop['人口総数'].astype(int)
```
