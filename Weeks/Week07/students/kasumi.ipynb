```python
# get center lat/lon
center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()

print(center_lat,center_lon)
```

```python
# loop and output multiple column values
for index, row in df.iterrows():
    print(row['name'],row['latitude'],row['longitude'])
```

```python
# make the map
m = folium.Map(location=[center_lat,center_lon], zoom_start=16)

# df ã‚’ãƒ«ãƒ¼ãƒ—ã—ã¦ãƒãƒ¼ã‚«ãƒ¼ã‚’ä½œã‚‹
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, 
                  popup='<div style="width:300px">'+row['description']+'</div>', 
                  tooltip=row['name'],
                  icon=folium.Icon(color=row['color']
                )).add_to(m)

# show the map
m
```

```python
# make the map
m = folium.Map(location=[center_lat,center_lon], 
               zoom_start=17,
               attr = 'å›½åœŸåœ°ç†é™¢')

# add basemaps
folium.TileLayer(name='2007å¹´ä»¥é™',tiles='https://cyberjapandata.gsi.go.jp/xyz/seamlessphoto/{z}/{x}/{y}.jpg',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1987å¹´ï½1990å¹´',tiles='https://cyberjapandata.gsi.go.jp/xyz/gazo4/{z}/{x}/{y}.jpg',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1984å¹´ï½1986å¹´',tiles='https://cyberjapandata.gsi.go.jp/xyz/gazo3/{z}/{x}/{y}.jpg',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1979å¹´ï½1983å¹´',tiles='https://cyberjapandata.gsi.go.jp/xyz/gazo2/{z}/{x}/{y}.jpg',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1974å¹´ï½1978å¹´',tiles='https://cyberjapandata.gsi.go.jp/xyz/gazo1/{z}/{x}/{y}.jpg',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1961å¹´ï½1969å¹´',tiles='https://cyberjapandata.gsi.go.jp/xyz/ort_old10/{z}/{x}/{y}.png',attr='å›½åœŸåœ°ç†é™¢').add_to(m)

# basemapãŒé¸ã¹ã‚‹ãƒœã‚¿ãƒ³ã‚’ãƒãƒƒãƒ—ã«è¶³ã™
folium.LayerControl().add_to(m)

# add points
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, popup='<div style="width:150px">'+row['description']+'</div>', tooltip=row['name'],icon=folium.Icon(color=row['color'])).add_to(m)

# show the map
m
```
