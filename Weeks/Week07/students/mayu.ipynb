```python
# get center lat/lonï¼ˆæ–°ã—ã„å¤‰æ•°ã‚’å…¥ã‚Œã‚‹ï¼‰

center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()
# Pointï¼šä¸­å¿ƒã‚’æŒ‡ã™ã€ã¨ã¦ã‚‚å¤§äº‹
print(center_lat,center_lon)
```

```python
# loop and output multiple column values
#è¦šãˆã‚‹ã—ã‹ãªã„ã€row(è¡Œ)
for index, row in df.iterrows():
    print(row['name'],row['latitude'],row['longitude'])
```

```python
# make the map
m = folium.Map(location=[center_lat,center_lon], zoom_start=13)

# df ã‚’ãƒ«ãƒ¼ãƒ—ã—ã¦ãƒãƒ¼ã‚«ãƒ¼ã‚’ä½œã‚‹
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, 
                  popup='<div style="width:180px">'+row['description']+'</div>', 
                  tooltip=row['name'],
                  icon=folium.Icon(color=row['color']
                )).add_to(m)

# show the map
m
```

```python
#2002å¹´ä»¥é™
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=56704
#1975ã€œ2001
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=806611
#1957~1974
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=712314
#1948å¹´ä»¥å‰
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=54315
```

```python
# make the map
#ãƒ‡ãƒ¼ã‚¿ã‚½ãƒ¼ã‚¹ã‚‚å¿…ãšè¨˜å…¥
m = folium.Map(location=[center_lat,center_lon], 
               zoom_start=13,
               attr = 'å›½åœŸåœ°ç†é™¢')

# add basemaps
folium.TileLayer(name='2002å¹´ä»¥é™',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=56704',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1975å¹´ã€œ2001å¹´',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=806611',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1957å¹´ã€œ1974å¹´',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=712314',attr='å›½åœŸåœ°ç†é™¢').add_to(m)
folium.TileLayer(name='1948å¹´ä»¥å‰',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=54315',attr='å›½åœŸåœ°ç†é™¢').add_to(m)

# basemapãŒé¸ã¹ã‚‹ãƒœã‚¿ãƒ³ã‚’ãƒãƒƒãƒ—ã«è¶³ã™
folium.LayerControl().add_to(m)

# add points
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, popup='<div style="width:150px">'+row['description']+'</div>', tooltip=row['name'],icon=folium.Icon(color=row['color'])).add_to(m)

# show the map
m
```
