```python
# library import
# pandas= データを集める、PDというあだ名をつけて使う
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
# import the csv file
df = pd.read_csv('mydataHW.csv')
```

```python
# get values for a single column
df['latitude']
```

```python
# put those values in a list
df['latitude'].to_list()
```

```python
# get center lat/lon（新しい変数を入れる）

center_lat = df['latitude'].mean()
center_lon = df['longitude'].mean()
# Point：中心を指す、とても大事
print(center_lat,center_lon)
```

```python
# loop and output multiple column values
#覚えるしかない、row(行)
for index, row in df.iterrows():
    print(row['name'],row['latitude'],row['longitude'])
```

```python
# make the map
m = folium.Map(location=[center_lat,center_lon], zoom_start=13)

# df をループしてマーカーを作る
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, 
                  popup='<div style="width:180px">'+row['description']+'</div>', 
                  tooltip=row['name'],
                  icon=folium.Icon(color=row['color']
                )).add_to(m)

# show the map
m
```

```python
#2002年以降
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=56704
#1975〜2001
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=806611
#1957~1974
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=712314
#1948年以前
#https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=54315
```

```python
# make the map
#データソースも必ず記入
m = folium.Map(location=[center_lat,center_lon], 
               zoom_start=13,
               attr = '国土地理院')

# add basemaps
folium.TileLayer(name='2002年以降',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=56704',attr='国土地理院').add_to(m)
folium.TileLayer(name='1975年〜2001年',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=806611',attr='国土地理院').add_to(m)
folium.TileLayer(name='1957年〜1974年',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=712314',attr='国土地理院').add_to(m)
folium.TileLayer(name='1948年以前',tiles='https://mapps.gsi.go.jp/map-lib-api/apiContentsView.do?specificationId=54315',attr='国土地理院').add_to(m)

# basemapが選べるボタンをマップに足す
folium.LayerControl().add_to(m)

# add points
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, popup='<div style="width:150px">'+row['description']+'</div>', tooltip=row['name'],icon=folium.Icon(color=row['color'])).add_to(m)

# show the map
m
```
