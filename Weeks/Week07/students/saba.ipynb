```python
m = folium.Map(location=[center_lat,center_lon], zoom_start=1)

# df をループしてマーカーを作る
for index, row in df.iterrows():
    latlon = [row['latitude'],row['longitude']]
    folium.Marker(latlon, 
                  popup='<div style="width:350px">'+row['description']+'</div>', 
                  tooltip=row['name'],
                  icon=folium.Icon(color=row['color']
                )).add_to(m)
m
```

```python
import folium

m = folium.Map(location=[center_lat, center_lon], zoom_start=1, attr='国土地理院')

# add basemaps
folium.TileLayer(name='1987年～1990年', tiles='http://tile.openstreetmap.org/{z}/{x}/{y}.png', attr='国土地理院').add_to(m)
folium.TileLayer(name='1984年～1986年', tiles='http://tile.openstreetmap.org/{z}/{x}/{y}.png', attr='国土地理院').add_to(m)
folium.TileLayer(name='1979年～1983年', tiles='http://tile.openstreetmap.org/{z}/{x}/{y}.png', attr='国土地理院').add_to(m)
folium.TileLayer(name='1974年～1978年', tiles='http://tile.openstreetmap.org/{z}/{x}/{y}.png', attr='国土地理院').add_to(m)
folium.TileLayer(name='1961年～1969年', tiles='http://tile.openstreetmap.org/{z}/{x}/{y}.png', attr='国土地理院').add_to(m)

# basemapが選べるボタンをマップに足す
folium.LayerControl().add_to(m)

# add points
for index, row in df.iterrows():
    latlon = [row['latitude'], row['longitude']]
    folium.Marker(latlon, popup='<div style="width:150px">' + row['description'] + '</div>', tooltip=row['name'], icon=folium.Icon(color='blue')).add_to(m)

# show the map
m
```
