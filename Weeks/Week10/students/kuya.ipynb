```python
import contextily as cx
```

```python
# change projection for your data to match basemap
gdf=gdf.to_crs(3857)
```

```python
# population map
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1)

# hide axes
ax.axis('off')

# title
ax.set_title('盛岡市の人口',fontsize=24);

# add basemap
cx.add_basemap(ax)
```

```python
# population map
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8)

# hide axes
ax.axis('off')

# title
ax.set_title('盛岡市の人口',fontsize=24);

# add basemap
cx.add_basemap(ax)
```

```python
# population map
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8)

# hide axes
ax.axis('off')

# title
ax.set_title('盛岡市の人口',fontsize=24);

# add basemap
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```

```python
gdf = gdf.sort_values(by='JINKO', ascending=False )
gdf
```

```python
# show top 10
gdf.head(10)
```

```python
fig = px.bar(gdf.head(10),
            y='JINKO',
            x='S_NAME',
            template='seaborn' # change this to see other styles
            )

fig.show()
```
