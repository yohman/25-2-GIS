```python
import contextily as cx
```

```python
gdf=gdf.to_crs(3857)
```

```python
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1)

# hide axes
ax.axis('off')

# title
ax.set_title('北海道札幌市中央区の人口',fontsize=24);

# add basemap
cx.add_basemap(ax)
```

```python
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.6)

# hide axes
ax.axis('off')

# title
ax.set_title('北海道札幌市中央区の人口',fontsize=24);

# add basemap
cx.add_basemap(ax)
```

```python
ax = gdf.plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8)

# hide axes
ax.axis('off')

# title
ax.set_title('北海道札幌市中央区の人口',fontsize=24);

# add basemap
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```

```python
ax = gdf.plot(figsize=(10,10),
         column='SETAI',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8)

# hide axes
ax.axis('off')

# title
ax.set_title('北海道札幌市中央区の世帯',fontsize=24);

# add basemap
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```

今回は今週末にこの街、北海道札幌市に旅行へ行くのでこの市を選びました。
mapを作ってみて、中央区の半分の面積は山に面しているため人口や世帯は片方に集中していることがわかりました。
ただ、人口の多い街のデータを見ると、地図の右上に位置する駅よりも山側に多く集まっている＆田畑が少なそうなので観光収入で生活している市民が多い街なのではと考えました。実際に行ってみて確かめてきます！
