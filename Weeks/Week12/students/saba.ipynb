```python
import geopandas as gpd
import pandas as pd
import contextily as ctx
import plotly.express as px
import contextily as cx
import numpy as np
import matplotlib.pyplot as plt

# set font
import matplotlib as mpl

# for Mac
mpl.rc('font',family='Hiragino Maru Gothic Pro')

# for PC
# mpl.rc('font',family='MS Gothic')
```

```python
# foreigners population map
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(20,20)) # 1 row, 2 columns, side by side


naganomesh.plot(ax = ax[0], # 
    column='T001100001',
         legend=True,
         legend_kwds={'shrink': 0.3},
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8,
         vmin=0,
         vmax=15000)

naganomesh.plot(ax = ax[1], # 
    column='T001100031',
         legend=True,
         legend_kwds={'shrink': 0.3},
         cmap='autumn_r',
         edgecolor='black',
         linewidth=0.1,
         alpha=0.8,
         vmin=0,
         vmax=15000)

# hide axes
ax[0].axis('off') # 
ax[1].axis('off') # 

# title
ax[0].set_title('2015 ',fontsize=24); # 
ax[1].set_title('2020 ',fontsize=24); # 

# Get the total bounds of the geometries
xmin, ymin, xmax, ymax = naganomesh.total_bounds

# Set the plot limits based on the total bounds
ax[0].set_xlim(xmin, xmax) # 
ax[0].set_ylim(ymin, ymax) #  
ax[1].set_xlim(xmin, xmax) # 
ax[1].set_ylim(ymin, ymax) # 

# 
plt.tight_layout()

# add basemap
cx.add_basemap(ax[0],source=cx.providers.Stamen.TonerLite) # 
cx.add_basemap(ax[1],source=cx.providers.Stamen.TonerLite) # 
```

```python
# create a new field
naganomesh['pop_change_2015_2020'] = np.nan
```

```python
# check!
naganomesh[['KEY_CODE','T001100001','T001100031','pop_change_2015_2020']].sample(5)
```

```python
# population map
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(12,12)) # nrows,ncols=1

# 2015-2020 population change
naganomesh.plot(ax = ax,
        column='pop_change_2015_2020',
        legend=True,
        legend_kwds={'shrink': 0.5},
        cmap='autumn_r',
        edgecolor='black',
        linewidth=0.1,
        alpha=0.8,)

# hide axes
ax.axis('off')

# title
ax.set_title('2015  2020 ',fontsize=24, pad=20);

# Get the total bounds of the geometries
xmin, ymin, xmax, ymax = naganomesh.total_bounds

# Set the plot limits based on the total bounds
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

# add basemap
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```
